```python
#importing libraries
import spacy
import re

#Giving input for the text and extracting bank names from the text
nlp = spacy.load('/content/drive/MyDrive/Chatbot/Rutva/bank_ner/')
amount=[]
def get_bank_currency(text):

    """
    returns banks name and currency in rand from the text
    
    """
    text=[text]
    string = str(text)
    doc = nlp(string)
    bank_extraction = [(ent.text, ent.label_) for ent in doc.ents]
    for txt in text:
            x = re.findall('[0-9]+' ' rand', str(txt))
            y = re.findall('[0-9]+rand', str(txt))
            if x:
                amount.append(x[0])
            elif y:
                amount.append(y[0])
            else:
                amount.append(re.findall('[0-9]+', str(text))[0])

    return bank_extraction, amount

sent = input("Enter Sentence: ")
print(get_bank_currency(sent))
```
