```python
#importing libraries
import pandas as pd
import nltk
import spacy
from spacy import tokenizer
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.tag import pos_tag
```

```python
#Sentences
text = ["My name is Brajkishore Prajapati .","Vineet Patel works at the bank.","Parth  is going to the bank.",
        "Diya Pancholi got a loan from the bank.", "Riya Patel has some money in the bank", "Rishi Patel has plenty of money in the bank.",
        "Jinal Thakkar doesn't wants to go to the bank.", "Nihar Thakker deposited a lot of money in the bank.", "Rushin Patel is a clerk at the African Bank .",
        "Khushali Sheth works in a bank as a secretary.", "Saumil Trivedi went to the bank and changed his currency.", "Brinda Nathwani will make the payment by bank transfer.",
        "According to John, the bank closes at 3 p.m.", "Mary keeps her jewels locked up in the bank.", "Jenny applied for a job with the Bank of Africa."]

str(text)
```

```python
#Loading package from spacy
nlp = spacy.load('en_core_web_sm')
```

```python
#Using spacy inserting text
doc = nlp(str(text))
print(doc)
```

```python
#NER 
x = [(X.text, X.label_) for X in doc.ents]
x
```

```python
#Creating DataFrame for Ner label == PERSON
PERSON_NER = [(X.text, X.label_) for X in doc.ents\
            if (X.label_ == 'PERSON' )]
          
dfn = pd.DataFrame(PERSON_NER)
dfn.rename(columns={0:"Names", 1:"NER_PERSON"}, inplace=True)
dfn
```

```python
#POS_tags
tagged = nltk.pos_tag(nltk.word_tokenize(str(text)))
tagged
```

```python
#NNP proper noun, singular – ‘Harrison’ & NNPS proper noun, plural – ‘Americans’ 
#Creating DataFrame for proper noun in POS
proper_noun = [word for word,pos in tagged \
            if (pos == 'NNP' or  pos == 'NNPS')]
dfp = pd.DataFrame(proper_noun)
dfp.rename(columns={0:"POS_PROPER_NOUN"}, inplace=True)
dfp
```

```python
#Creating DataFrame for sentences
data = pd.DataFrame(text)
data.rename(columns={0:"Name_Sentences"}, inplace=True)
data
```

```python
#Merging all the  DataFrames
df = pd.concat([data, dfn, dfp], axis=1)
df
```
